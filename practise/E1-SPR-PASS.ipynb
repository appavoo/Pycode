```python
shift_log = { "Karthik": {"I1": True, "I2": False, "I3": True}, "Alex": {"I4": True, "I5": True}, "Dinesh": {"I6": False, "I7": False, "I8": True} }
print(shift_log)

tested=0
passed=0

person={}

for supervisor,status in shift_log.items():
    person[supervisor]=0
    
    for tag,result in status.items():
               
        tested+=1
        if result:
            person[supervisor]+=1
            passed+=1
            print( f'passed in {passed}')
            
        else:
            print('nothing')
for name,count in person.items():
    print(f' the supervisor {name} has passed {count}')
    
    

print(f' total passed = {passed}')
print(f' total tested = {tested}')
print(f' supervisor = {person} and passed qty is {passed}')

            
    
    

```

```python
import random
full_choice=['r','p','s']


computer_score=0
user_score=0
won = []


while computer_score<4 and user_score<4:
    
    try:
       
        user_guess=input('Please enter r,p or s: ')
        
        full_choice.index(user_guess)
            #print(full_choice.index(user_guess))
        computer_guess=random.choice(full_choice)
        print(f'computer chose {computer_guess}')


        if user_guess== computer_guess:
            print('its a draw')
            print(f'computer score is {computer_score},user score is {user_score}')

        elif (user_guess=='r' and computer_guess=='s') or \
        (user_guess=='s' and computer_guess=='p') or \
        (user_guess=='p' and computer_guess=='r'):
            print('Adhira got a point')
            user_score+=1
            print(f'computer score is {computer_score},user score is {user_score}')

        elif (computer_guess=='r' and user_guess=='s') or \
        (computer_guess=='s' and user_guess=='p') or \
        (computer_guess=='p' and user_guess=='r'):
            print('computer got a point')
            computer_score+=1
            print(f'computer: {computer_score},You: {user_score}')
        
                  
    except ValueError:
        print('enter the correct letter')
    #finally:
        #print('enter either,p or s')

winner_score=max(computer_score,user_score)

if computer_score>user_score:
    winner= 'computer'
    won.append(winner)
    
else:
    winner= 'Adhira'
    won.append(winner)
  
    
    
print(f' the winner is {won} and the {won} has scored {winner_score}')
    
    
    
#print(f' the final score for computer is {computer_score} and user is {user_score}  and the winner is {}')
```
